```python
## Mouse Coords
##
## Left, Top, Width, Height
textScore = (604, 226, 10, 15)
## Left, Top, Width, Height
textGameOver = (258, 260, 50, 10)
## Left, Top, Width, Height
dinoBox = (70, 310, 50, 50)
## Left, Top, Width, Height
dinoTail = (75, 334, 5, 8)
## 
futureObsticles = (850,321,1000,345)
```

```python
xOffset = 102
xOffset2 = 0
xIncrement = 11
yOffset = 10
speed = 1
count = 0
countLast = 0
scoreCount = 0
timeout = 0
jumpCount = 0
jumping = False
doJump = False
```

```python
def f_stopMe():
    raise SystemExit("Stoping")
```

```python
def f_gameStart():
    # pyautogui.click(textGameOver)
    keyboard.press_and_release('space')     
    print("CLICK START")
```

```python
def f_jump():
    print("Jump")
    keyboard.press_and_release('space')
```

```python
def f_checkGameOver(saveImg):
    box = textGameOver
    
    savePath = "";
    if saveImg:
        savePath =  "images/game-over.png"
    
    im2 = f_grabImage(box, savePath)

    avg = numpy.average(im2, axis=0)
    avg = numpy.average(avg, axis=0)
    avg = numpy.average(avg, axis=0)
    
    if avg < 216 :
        print("===========")
        print(avg)
        print("Game Over")
        f_stopMe()
```

```python
def f_checkScore(saveImg):
    global xOffset2, scoreCount
    
    box = textScore
    
    savePath = "";
    if saveImg:
        savePath =  "images/score.png"
    
    im2 = f_grabImage(box, savePath)

    avg = numpy.average(im2, axis=0)
    avg = numpy.average(avg, axis=0)
    avg = numpy.average(avg, axis=0)
    
    if avg > 241 and scoreCount > 100:
        xOffset2 += xIncrement
        scoreCount = 0
        print(" ")
        print("======= INC : " + str(xOffset2))
        print(" ")
```

```python
def f_checkBox(saveImg):
    global jumping, jumpCount, doJump

    box = (dinoBox[0] + xOffset + xOffset2, dinoBox[1], dinoBox[2], dinoBox[3] - 10)
    
    savePath = "";
    if saveImg:
        savePath =  "images/boxes/"+ str(count) + "-box.png"
    
    im2 = f_grabImage(box, savePath)

    avg = numpy.average(im2, axis=0)
    avg = numpy.average(avg, axis=0)
    avg = numpy.average(avg, axis=0)
    
    print("B("+ str(count) +"): "+ str(avg))
    
    if jumping:
        jumpCount += 1
        if (jumpCount > 4 and avg > 230) or jumpCount > 6:
            jumping = False
            print("Cleared")
    
    if (avg < 241 and jumping == False) or doJump:
        jumping = True
        jumpCount = 0
        if f_checkTail(False) == False:
            jumping = False
            doJump = True
        else:
            doJump = False
            f_jump()
    
    
```

```python
def f_checkTail(saveImg):

    box = dinoTail
    
    savePath = "";
    if saveImg:
        savePath =  "images/dino-tail.png"
    
    im2 = f_grabImage(box, savePath)

    avg = numpy.average(im2, axis=0)
    avg = numpy.average(avg, axis=0)
    avg = numpy.average(avg, axis=0)
    
    print("T("+ str(count) +"): "+ str(avg))
    
    if avg > 210:
        return False
    
    return True
```

```python
def f_displayDino():
    box = dinoBox
    savePath =  "images/dino.png"
    f_grabImage(box, savePath)
    print(box)
    
```

```python
def f_grabImage(box, savePath):
    with mss.mss() as sct:
        # The screen part to capture
        monitor = {"top": box[1], "left": box[0], "width": box[2], "height": box[3]}

        # Grab the data
        img = sct.grab(monitor)
        ary = numpy.array(img)
        
        if len(savePath) > 1:
            mss.tools.to_png(img.rgb, img.size, output=savePath)
        
        return ary
    
```
